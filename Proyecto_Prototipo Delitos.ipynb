# (d)  El número promedio de delitos agrupados por tipo de delito y año y mes en la ciudad de México ordenados de mayor a menor

```python
D1 = df.groupby(['Año','Entidad','Bien jurídico afectado', 'Tipo de delito', 'Subtipo de delito', 'Modalidad']).agg({
    'Enero': 'mean',
    'Febrero': 'mean',
    'Marzo': 'mean',
    'Abril': 'mean',
    'Mayo': 'mean',
    'Junio': 'mean',
    'Julio': 'mean',
    'Agosto': 'mean',
    'Septiembre': 'mean',
    'Octubre': 'mean',
    'Noviembre': 'mean',
    'Diciembre': 'mean',
    'Anual': 'mean'
}).reset_index()
```

```python
D1
```

```python
D1_filtrado = D1[D1['Entidad'] == 'Ciudad de México']
```

```python
D1_ordenado = D1_filtrado.sort_values(by='Anual', ascending=False)
```

```python
D1_ordenado
```
